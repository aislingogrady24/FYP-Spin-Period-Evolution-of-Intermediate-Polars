```python
import numpy as np
import matplotlib.pyplot as plt
import astropy.units as u
import astropy.constants as c
import pandas as pd
```

```python
pi = np.pi
G = c.G
u0 = c.mu0
M1 = 0.5*c.M_sun
M2 = 0.7*c.M_sun
M3 = 1*c.M_sun
```

```python
AO_Piscium = pd.read_csv('AO_Piscium_Spin_Periods.csv')
print(AO_Piscium)
```

```python
BG_Canis_Minoris = pd.read_csv('BGCMi_Spin_Periods.csv')
print(BG_Canis_Minoris)
```

```python
DQ_Herculis = pd.read_csv('DQ_Herculis_Spin_Periods.csv')
print(DQ_Herculis)
```

```python
FO_Aquarii = pd.read_csv('FO_Aquarii_Spin_Periods.csv')
print(FO_Aquarii)
```

```python
aopsc_pspin = AO_Piscium.iloc[4,3]
bgcmi_pspin = BG_Canis_Minoris.iloc[3,4]
dqher_pspin = DQ_Herculis.iloc[4,3]
foaqr_pspin = FO_Aquarii.iloc[3,3]

omegaao = (2*pi)/aopsc_pspin
omegabg = (2*pi)/bgcmi_pspin
omegadq = (2*pi)/dqher_pspin
omegafo = (2*pi)/foaqr_pspin

print(aopsc_pspin, bgcmi_pspin, dqher_pspin, foaqr_pspin)
print(omegaao, omegabg, omegadq, omegafo)
```

```python
period = np.linspace(50, 1300, 1000)
omega = (2*pi)/period
Rco1 = ((G*M1)/omega**2)**(1/3)
Rco2 = ((G*M2)/omega**2)**(1/3)
Rco3 = ((G*M3)/omega**2)**(1/3)

plt.plot(period, Rco1, color='cornflowerblue', label='0.5M☉')
plt.plot(period, Rco2, color='mediumpurple', label='0.7M☉')
plt.plot(period, Rco3, color='deeppink', label='1M☉')
plt.xlabel('White Dwarf Spin Period (s)')
plt.ylabel('Corotation Radius (m)')
plt.vlines(aopsc_pspin,0, 1.75e8, linestyle='dashed')
plt.vlines(bgcmi_pspin,0, 1.75e8, linestyle='dashed')
plt.vlines(dqher_pspin,0, 1.75e8, linestyle='dashed')
plt.vlines(foaqr_pspin,0, 1.77e8, linestyle='dashed')
plt.text(720, 1.4e8, 'AO Psc')
plt.text(80, 0, 'DQ Her')
plt.text(920, 1.6e8, 'BG CMi')
plt.text(1100, 1e8, 'FO Aqr')
plt.legend()

plt.title('Corotation Radius vs Spin Period')
```

```python
M = 0.55*c.M_sun

pspin = np.linspace(0, 1000, 1000)*u.s
omega = (2*pi)/pspin
Rco = ((G*M)/omega**2)**(1/3)

B1 = 300*u.T  # 3 MG
Rm = 8.91e6*u.m
G = c.G
u0 = c.mu0
m_dot = 12.8E13*u.kg/u.s
alpha = 2*(pi**2)*(B1**4)*(Rm**12)
beta = G*M*u0**2
gamma = alpha/beta
Rusph = (gamma/(m_dot**2))**(1/7)
Rudisc1 = 0.52*Rusph.to(u.solRad)

B2 = 500*u.T  # 3 MG
Rm = 8.91e6*u.m
G = c.G
u0 = c.mu0
m_dot = 12.8E13*u.kg/u.s
alpha = 2*(pi**2)*(B2**4)*(Rm**12)
beta = G*M*u0**2
gamma = alpha/beta
Rusph = (gamma/(m_dot**2))**(1/7)
Rudisc2 = 0.52*Rusph.to(u.solRad)

B3 =100*u.T  # 3 MG
Rm = 8.91e6*u.m
G = c.G
u0 = c.mu0
m_dot = 12.8E13*u.kg/u.s
alpha = 2*(pi**2)*(B3**4)*(Rm**12)
beta = G*M*u0**2
gamma = alpha/beta
Rusph = (gamma/(m_dot**2))**(1/7)
Rudisc3 = 0.52*Rusph.to(u.solRad)

plt.plot(pspin, Rco.to(u.solRad), color='red', label='Corotation Radius')
plt.hlines(Rudisc1.value, 0, 1000, color='blue', label = 'Truncation Radius', linestyles='dashed')
plt.hlines(Rudisc2.value, 0, 1000, color='blue', linestyles='dashed')
plt.hlines(Rudisc3.value, 0, 1000, color='blue', linestyles='dashed')
plt.vlines(858.6, 0, 0.3, color='green', label = 'Current Spin Period', linestyles='dashed')
plt.text(150, 0.09, 'B = 1MG', fontsize=12)
plt.text(150, 0.17, 'B = 3MG', fontsize=12)
plt.text(150, 0.23, 'B = 5MG', fontsize=12)
plt.legend()
plt.xlabel('White Dwarf Spin Period (s)', fontsize=16)
plt.ylabel('Corotation Radius (Solar Radii)', fontsize=16)
plt.title('AO Piscium Corotation Radius Evolution', fontsize=18)
```
